```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

# =========================================================
# DATABASE SILVER
# =========================================================

spark.sql("CREATE DATABASE IF NOT EXISTS silver")

# =========================================================
# CLIENTE
# =========================================================

df_cliente = spark.table("bronze.Cliente")

df_cliente_silver = (
    df_cliente
    .dropDuplicates(["id_cliente"])
    .withColumn("nome_cliente", initcap(trim(col("nome_cliente"))))
    .withColumn("segmento", initcap(trim(col("segmento"))))
    .withColumn("cidade", upper(trim(col("cidade"))))
    .withColumn("estado", upper(trim(col("estado"))))
)

(
    df_cliente_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.Cliente")
)

# =========================================================
# CENTRO DE DISTRIBUICAO
# =========================================================

df_centro = spark.table("bronze.CentroDistribuicao")

df_centro_silver = (
    df_centro
    .dropDuplicates(["id_centro"])
    .withColumn("nome_centro", initcap(trim(col("nome_centro"))))
    .withColumn("cidade", upper(trim(col("cidade"))))
    .withColumn("estado", upper(trim(col("estado"))))
    .withColumn("capacidade_operacional", col("capacidade_operacional").cast("int"))
)

(
    df_centro_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.CentroDistribuicao")
)

# =========================================================
# VEICULO
# =========================================================

df_veiculo = spark.table("bronze.Veiculo")

df_veiculo_silver = (
    df_veiculo
    .dropDuplicates(["id_veiculo"])
    .withColumn("placa", upper(trim(col("placa"))))
    .withColumn("modelo", initcap(trim(col("modelo"))))
    .withColumn("tipo_veiculo", initcap(trim(col("tipo_veiculo"))))
    .withColumn("status_veiculo", upper(trim(col("status_veiculo"))))
    .withColumn("capacidade_kg", col("capacidade_kg").cast("int"))
    .withColumn("ano_fabricacao", col("ano_fabricacao").cast("int"))
)

(
    df_veiculo_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.Veiculo")
)

# =========================================================
# MOTORISTA
# =========================================================

df_motorista = spark.table("bronze.Motorista")

df_motorista_silver = (
    df_motorista
    .dropDuplicates(["id_motorista"])
    .withColumn("nome_motorista", initcap(trim(col("nome_motorista"))))
    .withColumn("cidade", upper(trim(col("cidade"))))
    .withColumn("estado", upper(trim(col("estado"))))
    .withColumn("categoria_cnh", upper(trim(col("categoria_cnh"))))
)

(
    df_motorista_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.Motorista")
)

# =========================================================
# ROTA
# =========================================================

df_rota = spark.table("bronze.Rota")

df_rota_silver = (
    df_rota
    .dropDuplicates(["id_rota"])
    .withColumn("cidade_origem", upper(trim(col("cidade_origem"))))
    .withColumn("estado_origem", upper(trim(col("estado_origem"))))
    .withColumn("cidade_destino", upper(trim(col("cidade_destino"))))
    .withColumn("estado_destino", upper(trim(col("estado_destino"))))
    .withColumn("distancia_km", col("distancia_km").cast("int"))
)

(
    df_rota_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.Rota")
)

# =========================================================
# TIPO CARGA
# =========================================================

df_tipo_carga = spark.table("bronze.TipoCarga")

df_tipo_carga_silver = (
    df_tipo_carga
    .dropDuplicates(["id_tipo_carga"])
    .withColumn("descricao_tipo_carga", initcap(trim(col("descricao_tipo_carga"))))
    .withColumn("categoria", initcap(trim(col("categoria"))))
)

(
    df_tipo_carga_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.TipoCarga")
)

# =========================================================
# ENTREGA
# =========================================================

df_entrega = spark.table("bronze.Entrega")

df_entrega_silver = (
    df_entrega
    .dropDuplicates(["id_entrega"])
    .withColumn("status_entrega", upper(trim(col("status_entrega"))))
    .withColumn("peso_carga_kg", col("peso_carga_kg").cast("decimal(10,2)"))
    .withColumn("valor_frete", col("valor_frete").cast("decimal(10,2)"))
    .withColumn("custo_combustivel", col("custo_combustivel").cast("decimal(10,2)"))
    .withColumn("tempo_estimado_horas", col("tempo_estimado_horas").cast("int"))
    .withColumn("tempo_real_horas", col("tempo_real_horas").cast("int"))
    .withColumn("quantidade_volumes", col("quantidade_volumes").cast("int"))
    .withColumn("avaria", col("avaria").cast("int"))
    .withColumn("entrega_no_prazo", col("entrega_no_prazo").cast("int"))
)

(
    df_entrega_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.Entrega")
)

print("Camada Silver criada com sucesso.")
```
